```python
import cv2
import pytesseract
from pytesseract import Output
import re
```

```python
img = cv2.imread('/home/user/Desktop/data science/DEEP LEARNING/data/ocr/invoice-sample.jpg')
```

```python
cv2.imshow('invoice',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_data = pytesseract.image_to_data(img,output_type=Output.DICT)
```

```python
print(img_data)
```

```python
print(img_data.keys())
```

```python
print(img_data['conf'])
```

```python

text_len = len(img_data['text'])
print(text_len)
```

```python
date_pattern =  '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
mail_pattern = '^[a-z0-9](\.?[a-z0-9]){5,}@example\.net$'
```

```python
for i in range(text_len):
 if img_data['conf'][i]>30:
     if re.match(mail_pattern,img_data['text'][i]): 
      x,y,w,h = img_data['left'][i],img_data['top'][i],img_data['width'][i],img_data['height'][i]
      cv2.rectangle(img,(x,y),(x+w,y+h),color=(0,255,0),thickness=5)
cv2.imshow('ocr',img)
cv2.waitKey(0)
cv2.destroyWindow()
```
